### **Custom Dataset Builder**

In order to load roboflow dataset in yolov7pytorch format, we created a custom Dataset class that helps us to integrate to pytorch ecosystem

```python
import os
import cv2
import numpy as np
import torch
from torch.utils.data import Dataset, DataLoader
import matplotlib.pyplot as plt


class DataSetBuilder(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.image_dir = os.path.join(root_dir, 'images')
        self.label_dir = os.path.join(root_dir, 'labels')
        self.image_list = os.listdir(self.image_dir)
        self.label_list = os.listdir(self.label_dir)

    def __len__(self):
        return len(self.image_list)

    def __getitem__(self, idx):
        img_name = os.path.join(self.image_dir, self.image_list[idx])
        label_name = os.path.join(self.label_dir, self.image_list[idx].replace(".jpg", ".txt"))

        image = cv2.imread(img_name)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Read bounding box coordinates and class labels from the label file
        with open(label_name, 'r') as file:
            labels = file.readlines()

        # Parse labels
        bboxes = []
        class_labels = []

        for label in labels:
            # Assuming the format is: class_index x_center y_center width height
            parts = label.split()
            class_idx = int(parts[0])
            x_center, y_center, width, height = map(float, parts[1:])
            bboxes.append([x_center, y_center, width, height])
            class_labels.append(class_idx)

        target = {}
        target["bboxes"] = torch.tensor(bboxes, dtype=torch.float32)
        target["class_labels"] = torch.tensor(class_labels, dtype=torch.int64)


        if self.transform:
            transformed = self.transform(image=image, bboxes=bboxes, class_labels=class_labels)
            image = transformed['image']
            target["bboxes"] = torch.tensor(transformed['bboxes'], dtype=torch.float32)
            target["class_labels"] = torch.tensor(transformed['class_labels'], dtype=torch.int64)

        return image, target
```

### **Custom collation function**
PyTorch's default collator cannot handle the varying sizes of bounding box lists directly, so we need a custom collation function that bypasses the attempt to stack the target dictionaries.

```python
from torch.utils.data.dataloader import default_collate


def collate_fn(batch):
    images = [item[0] for item in batch]
    targets = [item[1] for item in batch]

    images = default_collate(images)

    # No need to collate targets as they are already a list of dictionaries
    return images, targets
```
